```python
array_x = data[1:,1]
array_x
```

```python
array_y = data[1:,2]
array_y
```

```python
def linear_regression(array_x,array_y,epsilon,alfa):
    k = 0
    b = 0
    error = 0
    m = len(array_y)
    
    iteration = 0
    
    while(True):
        for i in range(m):
            error += pow((k*array_x[i] + b - array_y[i]),2)

        error = error / (2*m)

        k_ = 0
        b_ = 0

        for i in range(m):
            k_ += array_x[i]*(k*array_x[i] + b - array_y[i])
            b_ += k*array_x[i] + b - array_y[i]

        k = k - (alfa/m) * k_
        b = b - (alfa/m) * b_
        print("iteration is " + str(iteration) + " k is "+ str(k) +" b is "+ str(b) +" error is " + str(error))
        
        iteration += 1
        
        if( abs((alfa/m) * k_) < epsilon and abs((alfa/m) * b_ )< epsilon ):
            break

    return k,b
```

```python
koeffs = linear_regression(array_x,array_y,0.00001,0.3)
```
